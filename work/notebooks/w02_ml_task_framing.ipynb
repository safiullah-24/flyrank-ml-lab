## 1. My lane as an ML task (type)

**Task Type:** Ranking / Scoring

The **Refresh / Content Opportunity Scoring** lane is best framed as a **ranking and scoring** task. The goal is to assign each content page a priority score so that SEO specialists and content editors can prioritize which pages should be reviewed first. Rather than making a simple yes/no prediction, the system ranks pages according to their estimated refresh priority using multiple observed signals. This ranking supports editorial decision-making while keeping the final judgment with the human reviewer.

## 2. Target or proxy

The ideal output of this project is a **priority score** that ranks content pages according to how strongly they should be considered for review or refresh. Because the starter dataset does not contain an observed future outcome showing whether a refresh was actually successful, a direct target is not available.

For this framing exercise, **trend_direction** can be used as a **temporary proxy** to explore the problem. However, it is a defined label derived from **trend_pct**, not an independently observed future outcome. In a production system, this proxy should be replaced with an observed outcome collected after the prediction period so that the model learns from real results rather than a predefined rule.

## 3. Success metric

A suitable success metric for this project is **Precision@K** (for example, Precision@20). This metric measures how many of the top-ranked pages are genuinely good candidates for editorial review. Since reviewers can only examine a limited number of pages, evaluating the quality of the highest-ranked recommendations is more meaningful than measuring overall accuracy.

A good result would be a ranking that consistently places the most valuable refresh opportunities near the top, allowing editors to spend their time reviewing pages that are more likely to benefit from further investigation.

## 5. Why ML beats a fixed rule here

A simple rule-based approach, such as refreshing every page with low impressions or a "down" trend, would overlook many important factors that influence content performance. In practice, pages should be evaluated using multiple signals together, including search visibility, click-through rate, average position, engagement, freshness, content characteristics, and historical trends.

Machine learning is more suitable because it can learn patterns from the combined effect of these features instead of relying on a few manually defined thresholds. This produces a more consistent and evidence-based ranking of pages that deserve review while still supporting, rather than replacing, human decision-making. The final recommendations should therefore be treated as decision-support, not proof that refreshing a page will improve performance.
